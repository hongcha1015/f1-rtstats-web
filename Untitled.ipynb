```python
import pandas as pd

lap_times = pd.read_csv('csv_datasets/lap_times.csv')
drivers = pd.read_csv('csv_datasets/drivers.csv')
races = pd.read_csv('csv_datasets/races.csv')
```

```python
cl_id = 844
mv_id = 830
race_id = 1096
```

```python
# extract race
race_data = lap_times[lap_times['raceId'] == race_id]
```

```python
# extract drivers
cl_rd = race_data[race_data['driverId'] == cl_id]
mv_rd = race_data[race_data['driverId'] == mv_id]
```

```python
# ensure data is sorted the way we want to compare
cl_rd = cl_rd.sort_values("lap")
mv_rd = mv_rd.sort_values("lap")
```

```python
cl_rd[['lap', 'position', 'milliseconds']]
mv_rd[['lap', 'position', 'milliseconds']]
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

plt.figure(figsize = (10, 5))
plt.scatter(cl_rd['lap'], cl_rd['milliseconds'], label = "Charles Leclerc", c = 'red')
plt.scatter(mv_rd['lap'], mv_rd['milliseconds'], label = "Max Verstappen", c = 'blue')

def format_ms(x, _):
    minutes = int(x // 60000)
    seconds = int((x % 60000) // 1000)
    millis = int(x % 1000)
    return f"{minutes}:{seconds:02d}.{millis:03d}"

plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(format_ms))

plt.gcf().set_facecolor("#1e1e1e")
plt.gca().set_facecolor("#2b2b2b")

plt.ylabel("Lap Time (mm:ss.ms)", color = 'white')
plt.xlabel("Lap", color = 'white')
plt.title("Lap Time Comparison", color = 'white')

plt.xticks(np.arange(0, cl_rd['lap'].max(), 5))
ms_min = min(cl_rd['milliseconds'].min(), mv_rd['milliseconds'].min())
ms_max = max(cl_rd['milliseconds'].max(), mv_rd['milliseconds'].max())
plt.yticks(np.arange(ms_min, ms_max, 1000))
plt.tick_params(axis='both', colors='white')

plt.grid(True, alpha = 0.4)
plt.legend()
plt.show()
```
